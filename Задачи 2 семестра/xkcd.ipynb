```python
import requests
import json
from IPython.display import display, Image

def xkcd_comics(x):
    response = requests.get(f"https://xkcd.com/{x}/info.0.json")
    text = response.json()
    display(Image(f"{text['img']}")) 
    return f"Заголовок: {text['safe_title']}, Дата: {text['day']}.{text['month']}.{text['year']}, Подпись: {text['alt']}"
    
    
```
